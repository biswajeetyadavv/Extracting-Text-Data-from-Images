# Downloading images

```python
import os
import requests

def download_images(url_list, folder_name):
    # Create the folder if it doesn't exist
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
    
    for idx, url in enumerate(url_list):
        try:
            # Download the image
            response = requests.get(url)
            response.raise_for_status()  # Check if request was successful

            # Save the image with an appropriate file name
            file_name = os.path.join(folder_name, f'image_{idx+1}.jpg')
            with open(file_name, 'wb') as file:
                file.write(response.content)
            
            print(f"Image {idx+1} downloaded successfully!")
        
        except requests.exceptions.RequestException as e:
            print(f"Failed to download image {idx+1}: {e}")

# Usage:
# download_images(df['image_link'], 'images_folder')
```

```python
from PIL import Image, ImageFile
import os

# Allow PIL to load truncated images
ImageFile.LOAD_TRUNCATED_IMAGES = True

# Path to the image folder
image_folder = 'images_folder'

# List all image files
image_files = os.listdir(image_folder)
image_files = [f for f in image_files if f.endswith(('.png', '.jpg', '.jpeg'))]

# Load and preprocess images
def preprocess_image(image_path):
    try:
        img = Image.open(image_path)
        img = img.resize((224, 224))  # Resize
        img = img.convert('RGB')  # Convert to RGB
        return img
    except Exception as e:
        print(f"Error loading {image_path}: {e}")
        return None  # Return None if the image can't be processed

# Loop through images and preprocess
preprocessed_images = []
for image_file in image_files:
    image_path = os.path.join(image_folder, image_file)
    img = preprocess_image(image_path)
    if img:
        preprocessed_images.append(img)

print(f"Successfully loaded {len(preprocessed_images)} images.")
```

```python
def is_image_valid(image_path):
    try:
        img = Image.open(image_path)
        img.verify()  # This will raise an error if the image is corrupted
        return True
    except Exception as e:
        print(f"Corrupted image: {image_path} - {e}")
        return False

# Remove or skip corrupted images
valid_image_files = [f for f in image_files if is_image_valid(os.path.join(image_folder, f))]
print(f"Valid images: {len(valid_image_files)}")
```

```python
import pytesseract

def extract_text_from_image(image):
    # Use Tesseract to extract text
    text = pytesseract.image_to_string(image)
    return text

# Apply OCR to each preprocessed image and print the extracted text
for img in preprocessed_images:
    text = extract_text_from_image(img)
    print("Extracted Text:", text)
```

```python
import csv

# Example of saving extracted text into a CSV file
output_csv = 'extracted_values.csv'

with open(output_csv, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['image_file', 'extracted_text'])  # Headers
    
    for image_file, img in zip(image_files, preprocessed_images):
        text = extract_text_from_image(img)
        writer.writerow([image_file, text])

print(f"Extracted text saved to {output_csv}")
```
